```python
df = pd.read_csv('emi_dataset.csv', low_memory=False)
print("Dataset loaded!")
print("Shape:", df.shape)
```

```python
# pd.to_numeric with errors='coerce' converts bad values to NaN
# We count how many become NaN per column
for col in ['age', 'monthly_salary', 'bank_balance']:
    bad = pd.to_numeric(df[col], errors='coerce').isna().sum()
    print(f"{col}: {bad} non-numeric values")
```

```python
# Convert to numeric, coercing errors to NaN
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['monthly_salary'] = pd.to_numeric(df['monthly_salary'], errors='coerce')
df['bank_balance'] = pd.to_numeric(df['bank_balance'], errors='coerce')

print("Done! New dtypes:")
print(df[['age', 'monthly_salary', 'bank_balance']].dtypes)
```

```python
df['gender'] = df['gender'].str.strip().str.capitalize()
df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})

print(df['gender'].unique())
```

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

```python
# Check unique values in categorical columns
missing = df.isnull().sum()
missing = missing[missing > 0]
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage': missing_pct})
print(missing_df)
```

```python
# Numerical — fill with median
for col in ['age', 'monthly_salary', 'credit_score', 'bank_balance', 'emergency_fund']:
    df[col] = df[col].fillna(df[col].median())

# monthly_rent — 0 makes sense (owned house = no rent)
df['monthly_rent'] = df['monthly_rent'].fillna(0)

# education — fill with most common value
df['education'] = df['education'].fillna(df['education'].mode()[0])

print("Missing values remaining:")
print(df.isnull().sum().sum())
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Check data types and duplicates again after cleaning
print("Shape:", df.shape)
print("\nData types:")
print(df.dtypes)
print("\nMissing values:", df.isnull().sum().sum())
print("\nDuplicates:", df.duplicated().sum())
```

```python
df['age'] = df['age'].astype(int)
print(df['age'].dtype)
print(df['age'].head())
```

```python
# Convert all financial columns to int (after filling missing values)
df['college_fees'] = df['college_fees'].astype(int)
df['travel_expenses'] = df['travel_expenses'].astype(int)
df['groceries_utilities'] = df['groceries_utilities'].astype(int)
df['other_monthly_expenses'] = df['other_monthly_expenses'].astype(int)
df['school_fees'] = df['school_fees'].astype(int)
```

```python
X = df.drop(columns=['emi_eligibility', 'max_monthly_emi'])
y_class = df['emi_eligibility']
y_reg = df['max_monthly_emi']

print("Features shape:", X.shape)
print("Classification target:", y_class.unique())
print("Regression target stats:")
print(y_reg.describe())
```

```python
df['emi_eligibility'].unique()
```

```python
from sklearn.model_selection import train_test_split

# First split: 70% train, 30% temp
X_train, X_temp, y_class_train, y_class_temp, y_reg_train, y_reg_temp = train_test_split(
    X, y_class, y_reg, test_size=0.30, random_state=42, stratify=y_class
)

# Second split: 50% of temp = 15% val, 15% test
X_val, X_test, y_class_val, y_class_test, y_reg_val, y_reg_test = train_test_split(
    X_temp, y_class_temp, y_reg_temp, test_size=0.50, random_state=42, stratify=y_class_temp
)

print("Train:", X_train.shape)
print("Validation:", X_val.shape)
print("Test:", X_test.shape)
```

```python
import os

os.makedirs('data', exist_ok=True)

# Save cleaned full dataset
df.to_csv('data/cleaned_data.csv', index=False)

# Save splits
X_train.to_csv('data/X_train.csv', index=False)
X_val.to_csv('data/X_val.csv', index=False)
X_test.to_csv('data/X_test.csv', index=False)

y_class_train.to_csv('data/y_class_train.csv', index=False)
y_class_val.to_csv('data/y_class_val.csv', index=False)
y_class_test.to_csv('data/y_class_test.csv', index=False)

y_reg_train.to_csv('data/y_reg_train.csv', index=False)
y_reg_val.to_csv('data/y_reg_val.csv', index=False)
y_reg_test.to_csv('data/y_reg_test.csv', index=False)

print("All files saved in /data folder!")
```
